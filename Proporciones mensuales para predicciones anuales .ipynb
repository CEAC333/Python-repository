```python
#import libraries 
import numpy as np
import pandas as pd
```

```python
#import Data por industria
file= 'Dataframe_Exportaciones_mensuales_version2.csv'
datos=pd.read_csv(file,
                       header=0,
                       index_col=0,
                       sep=',',
                       parse_dates=False,
                       skip_blank_lines=True,
                 encoding='latin-1')
datos.iloc[:,:]
```

```python
#modificar año a float, y mes a números, hacer query para quitar "test_set"
datos['Año']=datos['Año'].values.astype(np.float)
datos.Mes=datos.Mes.map({'Enero':1,'Febrero':2,'Marzo':3,'Abril':4,'Mayo':5,'Junio':6,'Julio':7,'Agosto':8,'Septiembre':9
                               ,'Octubre':10,'Noviembre':11,'Diciembre':12})
datos=datos.query('(Año<2016) or (Año==2016 & Mes<8)').reset_index(drop=True)
```

```python
datos
```

```python
#para proporciones 1, con sumas mensuales de todos los años. sacas el monto mensual total y de ahí sacas proporciones:
primer_propo=datos.groupby(['Mes'],as_index=False,sort=False)['Monto'].sum()
total_exp=sum(primer_propo.Monto)
propociones_1=primer_propo.Monto.values/total_exp
df_propor_1=pd.DataFrame(propociones_1,columns=['proporciones_mensuales'],index=range(0,12))
df_propor_1=df_propor_1#.join(primer_propo.Mes)
df_propor_1
```

```python
#para proporciones 2, promedio de las proporciones de cada año:

segund_propo=datos.groupby(['Año','Mes'],as_index=False,sort=False).sum()

año_2001=segund_propo.query('Año==2001').reset_index(drop=True)
total_2001=sum(año_2001.Monto)
propo_2001=año_2001.Monto.values/total_2001
df_propo_2001=pd.DataFrame(propo_2001,columns=['Proporciones1'])

año_2002=segund_propo.query('Año==2002').reset_index(drop=True)
total_2002=sum(año_2002.Monto)
propo_2002=año_2002.Monto.values/total_2002
df_propo_2002=pd.DataFrame(propo_2002,columns=['Proporciones2'])

año_2003=segund_propo.query('Año==2003').reset_index(drop=True)
total_2003=sum(año_2003.Monto)
propo_2003=año_2003.Monto.values/total_2003
df_propo_2003=pd.DataFrame(propo_2003,columns=['Proporciones3'])

año_2004=segund_propo.query('Año==2004').reset_index(drop=True)
total_2004=sum(año_2004.Monto)
propo_2004=año_2004.Monto.values/total_2004
df_propo_2004=pd.DataFrame(propo_2004,columns=['Proporciones4'])

año_2005=segund_propo.query('Año==2005').reset_index(drop=True)
total_2005=sum(año_2005.Monto)
propo_2005=año_2005.Monto.values/total_2005
df_propo_2005=pd.DataFrame(propo_2005,columns=['Proporciones5'])

año_2006=segund_propo.query('Año==2006').reset_index(drop=True)
total_2006=sum(año_2006.Monto)
propo_2006=año_2006.Monto.values/total_2006
df_propo_2006=pd.DataFrame(propo_2006,columns=['Proporciones6'])

año_2007=segund_propo.query('Año==2007').reset_index(drop=True)
total_2007=sum(año_2007.Monto)
propo_2007=año_2007.Monto.values/total_2007
df_propo_2007=pd.DataFrame(propo_2007,columns=['Proporciones7'])

año_2008=segund_propo.query('Año==2008').reset_index(drop=True)
total_2008=sum(año_2008.Monto)
propo_2008=año_2008.Monto.values/total_2008
df_propo_2008=pd.DataFrame(propo_2008,columns=['Proporciones8'])

año_2009=segund_propo.query('Año==2009').reset_index(drop=True)
total_2009=sum(año_2009.Monto)
propo_2009=año_2009.Monto.values/total_2009
df_propo_2009=pd.DataFrame(propo_2009,columns=['Proporciones9'])

año_2010=segund_propo.query('Año==2010').reset_index(drop=True)
total_2010=sum(año_2010.Monto)
propo_2010=año_2010.Monto.values/total_2010
df_propo_2010=pd.DataFrame(propo_2010,columns=['Proporciones10'])

año_2011=segund_propo.query('Año==2011').reset_index(drop=True)
total_2011=sum(año_2011.Monto)
propo_2011=año_2011.Monto.values/total_2011
df_propo_2011=pd.DataFrame(propo_2011,columns=['Proporciones11'])

año_2012=segund_propo.query('Año==2012').reset_index(drop=True)
total_2012=sum(año_2012.Monto)
propo_2012=año_2012.Monto.values/total_2012
df_propo_2012=pd.DataFrame(propo_2012,columns=['Proporciones12'])

año_2013=segund_propo.query('Año==2013').reset_index(drop=True)
total_2013=sum(año_2013.Monto)
propo_2013=año_2013.Monto.values/total_2013
df_propo_2013=pd.DataFrame(propo_2013,columns=['Proporciones13'])

año_2014=segund_propo.query('Año==2014').reset_index(drop=True)
total_2014=sum(año_2014.Monto)
propo_2014=año_2014.Monto.values/total_2014
df_propo_2014=pd.DataFrame(propo_2014,columns=['Proporciones14'])

año_2015=segund_propo.query('Año==2015').reset_index(drop=True)
total_2015=sum(año_2015.Monto)
propo_2015=año_2015.Monto.values/total_2015
df_propo_2015=pd.DataFrame(propo_2015,columns=['Proporciones15'])
```

```python
df_propo2_pen=df_propo_2001.join(df_propo_2002).join(df_propo_2003).join(df_propo_2004).join(df_propo_2005).join(df_propo_2006)
df_propo2_pen=df_propo2_pen.join(df_propo_2007).join(df_propo_2008).join(df_propo_2009).join(df_propo_2010).join(df_propo_2011)
df_propo2_pen=df_propo2_pen.join(df_propo_2012).join(df_propo_2013).join(df_propo_2014).join(df_propo_2015)
```

```python
df_propor_2=pd.DataFrame(df_propo2_pen.mean(axis=1),columns=['proporciones_mensuales'])
```

```python
#convertir Dataframes  a CSV:

df_propor_1.to_csv('proporciones1.csv')
df_propor_2.to_csv('proporciones2.csv')
```
