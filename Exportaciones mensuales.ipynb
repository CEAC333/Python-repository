```python
#import libraries 
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
```

```python
#import Data
file= 'exportacionesmen.csv'
datos=pd.read_csv(file,
                       header=0,
                       sep=',',
                       parse_dates=False,
                       skip_blank_lines=True,
                 encoding='latin-1')
datos
```

```python
#limpiar base de datos
datos=datos.replace('N/D',0)
datos2=datos.iloc[:,2:213]
datos3=datos.iloc[:,0:2]
datos2=datos2.astype(float)
datosc=datos3.join(datos2)

index= datosc.index.isin([5,11,17,23,29,35,41,47,48])
datosgroupbycountry=datosc[index]
datosgroupbycountry.to_csv('Groupexportdataindustry.csv')
datosc=datosc[~index]

datosgroupbycountry
```

```python
#separar paises compradores y los que se agruparan 
datosq=datosc.iloc[0:30,:]
datosp=datosc.iloc[30:,:]
datosp
```

```python
#group by industria para los paises que se convertirán en el 7° "pais" vendedor 
datosp=datosp.groupby('Industria').sum()
datosp
```

```python
#crear DataFrame del nuevo grupo
Industrias_Manufactureras=pd.Series(datosp.iloc[3,:])
Industrias_Extractivas=pd.Series(datosp.iloc[2,:])
Agricultura_y_Silvicultura=pd.Series(datosp.iloc[0,:])
Servicios_y_Productos_no_Clasificados=pd.Series(datosp.iloc[4,:])
Ganadería_Apicultura_Caza_Pesca=pd.Series(datosp.iloc[1,:])

data2 = [Industrias_Manufactureras,Industrias_Extractivas,Agricultura_y_Silvicultura,Servicios_y_Productos_no_Clasificados,
        Ganadería_Apicultura_Caza_Pesca]

data2=pd.DataFrame(data2,index=np.arange(5))
data2
```

```python
#data frame columnas str
d={'0' : 'Industrias Manufactureras', '1' : 'Industrias Extractivas', '2' : 'Agricultura y Silvicultura',
  '3' : 'Servicios y Productos no Clasificados', '4' : 'Ganadería, Apicultura, Caza, Pesca'}
a=pd.Series(d)

h={'Pais_comprador':pd.Series('otros', index=[0,1,2,3,4]),'industria':a}
w=pd.DataFrame(h)

w.iloc[0:5,1]=w.values[5:10,1]
data1=w.iloc[0:5,:]
columns=list(data1.columns.values)
columns[1]='Industria'
data1.columns=columns
data1
```

```python
#crear data frame del 7° pais vendedor
sept=data1.join(data2)
sept
```

```python
#crear data frame final 
datafinal=datosq.append(sept, ignore_index=True)
datafinal
```

```python
#convertir a CSV
datafinal.to_csv('DataframeExportacionesmensuales.csv')
```
