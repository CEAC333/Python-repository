```python
#import libraries 
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
from pylab import *
```

```python
#import Data
file= 'empleo_TamEmp.csv'
datos=pd.read_csv(file,
                       header=0,
                       sep=',',
                       parse_dates=False,
                       skip_blank_lines=True,
                 encoding='latin-1')
datos.head()
```

```python
#limpiar base datos
datos=datos.iloc[:,0:14]
columns=list(datos.columns.values)
columns[1]='Tamaño_empresa'
datos.columns=columns
datos=datos.replace('N/D',0)
datos2=datos.iloc[:,2:15]
datos3=datos.iloc[:,1:2]
datos2=datos2.astype(float)
datosc=datos3.join(datos2)
datosc
```

```python
#agrupar por tamaño de empresa

datosc=datosc.groupby('Tamaño_empresa').sum()
datosc
```

```python
index= datosc.index.isin(['División económica','Tamaño de empresa'])
datosc=datosc[~index]
datosc
```
