```python
#import libraries 
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
from pylab import *
```

```python
#import Data
file= 'Empleo.csv'
datos=pd.read_csv(file,
                       header=0,
                       sep=',',
                       parse_dates=False,
                       skip_blank_lines=True,
                 encoding='latin-1')
datos
```

```python
#limpiar base de datos
datos2=datos.iloc[:,2:21]
datos3=datos.iloc[:,0:2]
datos2=datos2.astype(float)
datosc=datos3.join(datos2)

index= datosc.index.isin([4,9,14,19,24,29,34,39,40])
datosc=datosc[~index]
datosc=datosc.iloc[:,0:20]

columns=list(datosc.columns.values)
columns[1]='tipo_empleo'
datosc.columns=columns
```
