```python
#import libraries 
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
from pylab import *
```

```python
#import Data por industria
file= 'DataframeExportacionesmensuales.csv'
datos1=pd.read_csv(file,
                       header=0,
                       index_col=0,
                       sep=',',
                       parse_dates=False,
                       skip_blank_lines=True,
                 encoding='latin-1')
datos1
```

```python
#import Data por seccion
file= 'DataframeExportmenporseccion.csv'
datos2=pd.read_csv(file,
                       header=0,
                       index_col=0,
                       sep=',',
                       parse_dates=False,
                       skip_blank_lines=True,
                 encoding='latin-1')
datos2.head(10)
```

```python
#group by pais_comprador 
datosporpais=datos1.groupby('Pais_comprador',as_index=True).sum()

datosporpais_AN=datosporpais.iloc[2,:]
datosporpais_resto=datosporpais.iloc[(0,1,3,4,5,6),:]
datosporpais_AOO=datosporpais.iloc[(0,5,6),:] #africa ,oceania y otros 
datosporpais
```

```python
%matplotlib inline
```

```python
#gráfica exportaciones por pais comprador(todos los paises)
datosporpais.transpose().plot(title='Exportaciones',figsize=(14,8))
```

```python
#se aprecian los cambios del 2008 y después de las elecciones presidenciales de EU(trump)
#gráfica exportaciones por pais comprador (américa del norte a parte)
datosporpais_AN.plot(title='Exportaciones a américa del norte',figsize=(14,8))
datosporpais_resto.transpose().plot(figsize=(14,8))
datosporpais_AOO.transpose().plot(figsize=(14,8))

#se aprecian los cambios del 2008 y después de las elecciones presidenciales de EU(trump)
```

```python
#Correlaciones entre paises
correlaciones=pd.DataFrame(np.corrcoef(datosporpais))
columnas=['Africa','América Latina','América del Norte','Asia','Europa','Oceanía','otros']
correlaciones.columns=columnas

corr = datosporpais.transpose().corr()
fig,ax = plt.subplots(figsize=(9, 9))
ax.matshow(corr)
plt.xticks(range(len(corr.columns)), corr.columns);
plt.yticks(range(len(corr.columns)), corr.columns);
correlaciones
```

```python
#gráfica de pastel (Exportaciones por país comprador ) /desde enero 2000 hasta diciembre 2008
sumas=datosporpais.iloc[:,:100].sum(axis=1)
suma=sumas.sum()
propor=sumas/suma

# make a square figure and axes
figure(1, figsize=(6,6))
ax = axes([0.1, 0.1, 0.8, 0.8])

# The slices will be ordered and plotted counter-clockwise.
labels = ['Otros','América Latina','América del Norte','Asia','Europa','.','.']
fracs = propor

pie(fracs,labels=labels,
                autopct='%1.1f%%', shadow=True, startangle=90,radius=1.05,pctdistance =.7)
                # The default startangle is 0, which would start
                # the Frogs slice on the x-axis.  With startangle=90,
                # everything is rotated counter-clockwise by 90 degrees,
                # so the plotting starts on the positive y-axis.

title('Exportaciones por país comprador (2000-2008)', bbox={'facecolor':'0.8', 'pad':5})

show()
```
