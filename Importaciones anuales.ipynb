```python
#import libraries 
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
```

```python
#import Data
file= 'importaciones.csv'
datos=pd.read_csv(file,
                       header=0,
                       sep=',',
                       parse_dates=False,
                       skip_blank_lines=True,
                 encoding='latin-1')
datos
```

```python
#limpiar base de datos
datos=datos.replace('N/D',0)
datos2=datos.iloc[:,2:21]
datos3=datos.iloc[:,0:2]
datos2=datos2.astype(float)
datosc=datos3.join(datos2)

index= datosc.index.isin([6,13,20,27,34,41,48,55,62,69,76,83,84])
datosgroupbycountry=datosc[index]
datosc=datosc[~index]
datosc=datosc.iloc[:,0:20]

columns=list(datosc.columns.values)
columns[0]='Pais_vendedor'
columns[1]='industria'
datosc.columns=columns

columnsg=list(datosgroupbycountry.columns.values)
columnsg[0]='Pais_vendedor'
columnsg[1]='industria'
datosgroupbycountry.columns=columnsg

datosc.head()
```

```python
#separar 6 paises vendedores y los que se convertirán en el 7° "país"
datosq=datosc.iloc[0:42,:]
index= datosq.index.isin([28,29,30,31,32,33])
datosnodisp=datosq[index]
datosq=datosq[~index]
datosq
```

```python
#group by industria para los paises que se convertirán en el 7° "pais" vendedor 
datosp=datosc.iloc[42:72,:]
datosp=datosp.groupby('industria').sum()
#k=pd.DataFrame(list(['otros','otros','otros','otros','otros','otros']))
#datosp.insert(loc=0,column='Pais_vendedor',value=k)
datosp
```

```python
#sumar valores "no disponible" con el "groupby industry" 
Industrias_Manufactureras=pd.Series(datosp.iloc[4,:])
Industrias_Extractivas=pd.Series(datosp.iloc[3,:])
Agricultura_y_Silvicultura=pd.Series(datosp.iloc[1,:])
Servicios_y_Productos_no_Clasificados=pd.Series(data=(datosp.iloc[5,:]+datosnodisp.iloc[3,2:20]))
Ganadería_Apicultura_Caza_Pesca=pd.Series(datosp.iloc[2,:])
blank=pd.Series(datosp.iloc[0,:])
```

```python
#crear data frame de valores sumados 
data2 = [Industrias_Manufactureras,Industrias_Extractivas,Agricultura_y_Silvicultura,Servicios_y_Productos_no_Clasificados,
        Ganadería_Apicultura_Caza_Pesca,blank]

data2=pd.DataFrame(data2,index=np.arange(6))
data2
```

```python
#data frame columnas str
d={'0' : 'Industrias Manufactureras', '1' : 'Industrias Extractivas', '2' : 'Agricultura y Silvicultura',
  '3' : 'Servicios y Productos no Clasificados', '4' : 'Ganadería, Apicultura, Caza, Pesca', '5' : '( Blank )'}
a=pd.Series(d)

h={'Pais_vendedor':pd.Series('otros', index=[0,1,2,3,4,5]),
   'industria':a}
w=pd.DataFrame(h)

w.iloc[0:6,1]=w.values[6:12,1]
data1=w.iloc[0:6,:]
data1
```

```python
#crear data frame del 7° pais vendedor
sept=data1.join(data2)
sept
```

```python
#crear data frame final 
datafinal=datosq.append(sept, ignore_index=True)
datafinal
```

```python
#convertir a CSV
datafinal.to_csv('DataframeImportaciones.csv')
```
