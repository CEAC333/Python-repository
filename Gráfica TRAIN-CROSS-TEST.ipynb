```python
#import libraries 
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
```

```python
%matplotlib inline
```

```python
#import Data exportaciones
file= 'Dataframe_Exportaciones_mensuales_version2.csv'
datos=pd.read_csv(file,
                       header=0,
                       index_col=0,
                       sep=',',
                       parse_dates=False,
                       skip_blank_lines=True,
                 encoding='latin-1')
datos.iloc[:,:]
```

```python
#import Data por industria
file= 'Dataframe_Importaciones_mensuales_version2.csv'
datos_imp=pd.read_csv(file,
                       header=0,
                       index_col=0,
                       sep=',',
                       parse_dates=False,
                       skip_blank_lines=True,
                 encoding='latin-1')
datos_imp.iloc[:,:]
```

```python
#Agrupar Datos para obtener serie mensual sin segmentos, Enero2000-Julio 2017 ##exportaciones

serie_total=datos.groupby(['Año','Mes'],as_index=False,sort=False).sum()
serie_total['Año']=serie_total['Año'].values.astype(str)
fechas=serie_total.Año+'-'+serie_total.Mes
d={'Fecha':fechas,'Monto':serie_total.Monto}
serie_total=pd.DataFrame(d)
serie_total=serie_total.set_index('Fecha')
serie_total.plot(title='Serie de Exportaciones totales',figsize=(14,8),grid=True,colormap='ocean')
```

```python
#Agrupar Datos para obtener serie mensual sin segmentos, Enero2000-Julio 2017 ##Importaciones

serie_total_import=datos_imp.groupby(['Año','Mes'],as_index=False,sort=False).sum()
serie_total_import['Año']=serie_total_import['Año'].values.astype(str)
fechas_imp=serie_total_import.Año+'-'+serie_total_import.Mes
d_imp={'Fecha':fechas_imp,'Monto':serie_total_import.Monto}
serie_total_import=pd.DataFrame(d_imp)
serie_total_import=serie_total_import.set_index('Fecha')
serie_total_import.plot(title='Serie de Importaciones totales',figsize=(14,8),grid=True,colormap='Blues_r')
```

```python
ax = serie_total_import.plot(title='Serie de Importaciones totales',figsize=(14,8),grid=True,colormap='Blues_r')
ymin, ymax = ax.get_ylim()
dates=['180','198']
ax.vlines(x=dates, ymin=ymin, ymax=ymax-1, color='red')
```
